#### Bubble Sort
- 인접한 두 개의 원소를 비교하며 자리를 계속 교환하는 방식 (O(n^2))

- 정렬 과정
    - 첫 원소부터 인접한 원소끼리 계속 자리를 교환하며 맨 마지막 자리까지 이동.
    - 한 단계가 끝나면 가장 큰 원소가 마지막 자리로 정렬.

````python
def BubbleSort(a, N): # 정렬할 List, 원소 수 
    for i in range(N-1, 0, -1):
        for j in range(0,i):
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1],a[j]
    return(a)
Lst = [1,6,4,3,2,6,8,2,8,9,2,7,6,4]
print(BubbleSort(Lst,14))
````

### 검색 
저장되어 있는 자료 중에서 원하는 항목을 찾는 작업<p>
종류
- Sequential search
    - 가장 간단. 직관적. 첫 원소부터 순서대로 검색 대상과 키 값이 같은 원소가 있는지 비교
    - 찾으면 인덱스 반환, 마지막까지 이를 때까지 대상을 찾지 못하면 검색 실패
- binary search
    - 자료의 가운데 항목 키 값과 비교하고 다음 검색의 위치를 결정하고 검색을 계속 진행.
    - 자료가 정렬된 상태여야 한다.
    - 자료 중앙원소 선택 -> 목표값과 비교 -> 크다면 오른쪽으로 작다면 왼쪽으로
- hash
    

````python
def binarySearch(a, N ,key):
    start = 0
    end = N-1
    while start <= end:
        mid = (start + end )// 2
        if a[mid] == key:
            return True # 검색 성공
        elif a[mid] > key :
            end = mid -1
        else : 
            start = mid +1 
    return False #실패
aa = [1, 3, 5, 6, 7, 8 ,9 ,10 ,11 ,22 ,34 ,567, 1123]
N = len(aa)
binarySearch(aa,N, 3)
````

#### Selection Sort

주어진 자료 중 가장 작은 값의 원소부터 차례대로 선택하여 위치 교환.


과정
- 최소값 찾기
- 맨 앞 위치한 값과 교환
- 맨 처음 위치 제외, 이를 반복함 O(n^2)

- 잘 사용하면 정리 후 K 번쨰로 작은 원소를 찾을 수 있음

````python
def selectionSort(a, N ):
    for i in range(N-1):
        minidx = i
        for j in range(i+1,N):
            if a[minidx] > a[j]:
                minidx = j
        a[i], a[minidx] = a[minidx],a[i]

B = [456,4, 5,6,1,7, 23,12 ,4, 65]
N = len(B)
selectionSort(B, N)
print(B)
````

````python
def BruteFoce(pt, txt): 
    # pt 는 패턴
    # txt 는 텍스트
    i = 0 # txt 의 인덱스
    j = 0 # pt의 인덱스
    while i < N and j < M : # N은 텍스트의 len, M은 패턴의 len
        if txt[i] != pt[j]:
            i = i - j  # 텍스트 인덱스 = 텍스트 인덱스 - pt 인덱스
            j = -1 # 아래에서 1개 증가할 예정이라서 -1임
        i = i + 1  # 하나 증가요 
        j = j + 1  # 하나 증가요
    if j == M:
        return 1 # 성공
    else:
        return -1 # 실패
````

#### DP - Dynamic Programming
- 동적 계획 알고리즘은 그리디 알고리즘과 같이 최적화 문제를 해결
- 작은 부분 문제 모두 해결 -> 그 해를 이용 큰 부분 문제 해결 -> 최종적으로 해결
ej. 피보나치 dp 적용
- 1) 부분 문제로 분할
    - F(n) = F(n-1) + F(n-2)
    - F(2) = F(1) + F(0)
    - F(n)은 암튼...
- 2) 부분 문제로 나누는 일 끝냈다면 작은 부분 문제부터
- 3) 결과는 테이블에 저장, 부분 문제의 해를 이용하여 상위 해 도출


````python
def fibo_dp(n):
    f = [0] * (n+1)
    f[0] = 0
    f[1] = 1
    for i in range(2, n+1):
        f[i] = f[i-1] + f[i-2]
    return f[n]

print(fibo_dp(18000)) # 와 압도적으로 빠르다!
````

#### DFS (깊이우선탐색)
- 비선형구조인 그래프 구조는 모든 자료 탐색이 중요할지도 몰라요
- DFS / BFS
- DFS::
    - 시작 정점의 한 방향으로 갈 수 있는 경로가 있는 곳까지 깊이 탐색해 가다가 더 이상 갈 곳이 없게 되면, 가장 마지막에 만났던 갈림길 간선이 있는 정점으로 되돌아와서 다른 방향의 정점으로 탐색을 계속 반복하여 결국 모든 정점을 방문하는 순회방법
    - 가장 마지막에 만났던 갈림길의 정점으로 되돌아가서 다시 깊이 우선 탐색을 반복해야 하므로 후입선출 구조의 스택 사용
##### DFS 알고리즘
   - 1) 시작 정점 v를 결정하여 방문
   - 2) 정점 v에 인접한 정점 중에서
        - A) 방문하지 않은 정점 w가 있으면 정점 v를 스택에 push 하고 정점w를 방문.
        - B) w를 v로 하여 다시 2를 반복
        - C) 방문하지 않은 정점이 없으면, 탐색의 방향을 바꾸기 위해서 스택을 pop하여 받은 마지막 방문 정점을 v로 하여 다시 2)를 반복한다.
   - 3) 스택이 공백이 될 때 까지 2) 반복
